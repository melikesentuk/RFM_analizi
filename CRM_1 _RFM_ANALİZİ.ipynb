 ðŸ”´ GÃ¶rev 2: RFM Metriklerinin HesaplanmasÄ±

 ðŸ”´AdÄ±m 1:  MÃ¼ÅŸteri Ã¶zelinde Recency, Frequency ve Monetary metriklerini hesaplayÄ±nÄ±z

```python
# son satÄ±n alma tarihini bulalÄ±m -
df["LAST_ORDER_DATE"].max()
```

```python
# +2 olarak analiz tarihini belirleyelim
analiz_date = dt.datetime(2021,6,1)
```

```python
analiz_date
```

```python
RFM = pd.DataFrame()  # boÅŸ bir dataframe olÅŸturalÄ±m
```

```python
RFM["CUSTOMERID"] = df["MASTER_ID"]
```

```python
#  mÃ¼ÅŸterinin son satÄ±n almasÄ± Ã¼zerinden ne kadar zaman geÃ§ti
RFM["Recency"] = (analiz_date - df["LAST_ORDER_DATE"]).astype("timedelta64[D]")
```

```python
# mÃ¼ÅŸterinin toplam satÄ±nalma sayÄ±sÄ± FREQUENcy
RFM["FREQUENCY"] = df["ORDER_NUM"]
```

```python
# MONETARY mÃ¼ÅŸterinin top lam harcadÄ±ÄŸÄ± tutar
RFM["MONETARY"] = df["TOTAL_NUM"]
```

```python
RFM.head()
```

ðŸ”´ AdÄ±m 1: Recency, Frequency ve Monetary metriklerini qcut yardÄ±mÄ± ile 1-5 arasÄ±nda skorlara Ã§evirelim

```python
RFM["Recency_score"] = pd.qcut(RFM["Recency"], 5, labels=[5,4,3,2,1])  # son sipariÅŸ tarihinden ne kadar az sÃ¼re geÃ§tiyse o gÃ¼ncel bir mÃ¼ÅŸteri o yÃ¼zden label kÄ±smÄ± 5 den 1 e doÄŸru
```

```python
RFM["Frequency_score"] = pd.qcut(RFM["FREQUENCY"].rank(method="first"), 5, labels= [1,2,3,4,5])  # frequency deÄŸerleri Ã§ok iÃ§ iÃ§e geÃ§tiÄŸi iÃ§in rank method kullanÄ±ldÄ±
```

```python
RFM["Monetary_Score"] = pd.qcut(RFM["MONETARY"], 5, labels=[1,2,3,4,5])
```

ðŸ”´  recency_score ve frequency_scoreâ€™u tek bir deÄŸiÅŸken olarak ifade ediniz ve RF_SCORE olarak kaydet

```python
RFM["RF_SCORE"] = RFM["Recency_score"].astype(str) + RFM["Frequency_score"].astype(str)  # stringe Ã§evirip bir araya getirelim
```

```python
RFM.head()
```

RF Skorunun Segment Olarak TanÄ±mlanmasÄ±

ðŸ”´ AdÄ±m 1: OluÅŸturulan RF skorlarÄ± iÃ§in segment tanÄ±mlamalarÄ± yap

```python
seg_map = {
        r'[1-2][1-2]': 'hibernating',
        r'[1-2][3-4]': 'at_Risk',
        r'[1-2]5': 'cant_loose',
        r'3[1-2]': 'about_to_sleep',
        r'33': 'need_attention',
        r'[3-4][4-5]': 'loyal_customers',
        r'41': 'promising',
        r'51': 'new_customers',
        r'[4-5][2-3]': 'potential_loyalists',
        r'5[4-5]': 'champions'
}
```

```python
# rf skor deÄŸerlerinin oluÅŸturduÄŸumuz seg map deÄŸerlerine gÃ¶re deÄŸiÅŸtirelim
RFM['segment'] = RFM['RF_SCORE'].replace(seg_map, regex=True)
```

```python
RFM.head()
```

ðŸ”´ AdÄ±m 1: Segmentlerin recency, frequnecy ve monetary ortalamalarÄ±nÄ± incele

```python
RFM.groupby("segment").agg({"Recency":"mean", "FREQUENCY": "mean", "MONETARY": "mean"})
```
