 🔴 Görev 2: RFM Metriklerinin Hesaplanması

 🔴Adım 1:  Müşteri özelinde Recency, Frequency ve Monetary metriklerini hesaplayınız

```python
# son satın alma tarihini bulalım -
df["LAST_ORDER_DATE"].max()
```

```python
# +2 olarak analiz tarihini belirleyelim
analiz_date = dt.datetime(2021,6,1)
```

```python
analiz_date
```

```python
RFM = pd.DataFrame()  # boş bir dataframe olşturalım
```

```python
RFM["CUSTOMERID"] = df["MASTER_ID"]
```

```python
#  müşterinin son satın alması üzerinden ne kadar zaman geçti
RFM["Recency"] = (analiz_date - df["LAST_ORDER_DATE"]).astype("timedelta64[D]")
```

```python
# müşterinin toplam satınalma sayısı FREQUENcy
RFM["FREQUENCY"] = df["ORDER_NUM"]
```

```python
# MONETARY müşterinin top lam harcadığı tutar
RFM["MONETARY"] = df["TOTAL_NUM"]
```

```python
RFM.head()
```

🔴 Adım 1: Recency, Frequency ve Monetary metriklerini qcut yardımı ile 1-5 arasında skorlara çevirelim

```python
RFM["Recency_score"] = pd.qcut(RFM["Recency"], 5, labels=[5,4,3,2,1])  # son sipariş tarihinden ne kadar az süre geçtiyse o güncel bir müşteri o yüzden label kısmı 5 den 1 e doğru
```

```python
RFM["Frequency_score"] = pd.qcut(RFM["FREQUENCY"].rank(method="first"), 5, labels= [1,2,3,4,5])  # frequency değerleri çok iç içe geçtiği için rank method kullanıldı
```

```python
RFM["Monetary_Score"] = pd.qcut(RFM["MONETARY"], 5, labels=[1,2,3,4,5])
```

🔴  recency_score ve frequency_score’u tek bir değişken olarak ifade ediniz ve RF_SCORE olarak kaydet

```python
RFM["RF_SCORE"] = RFM["Recency_score"].astype(str) + RFM["Frequency_score"].astype(str)  # stringe çevirip bir araya getirelim
```

```python
RFM.head()
```

RF Skorunun Segment Olarak Tanımlanması

🔴 Adım 1: Oluşturulan RF skorları için segment tanımlamaları yap

```python
seg_map = {
        r'[1-2][1-2]': 'hibernating',
        r'[1-2][3-4]': 'at_Risk',
        r'[1-2]5': 'cant_loose',
        r'3[1-2]': 'about_to_sleep',
        r'33': 'need_attention',
        r'[3-4][4-5]': 'loyal_customers',
        r'41': 'promising',
        r'51': 'new_customers',
        r'[4-5][2-3]': 'potential_loyalists',
        r'5[4-5]': 'champions'
}
```

```python
# rf skor değerlerinin oluşturduğumuz seg map değerlerine göre değiştirelim
RFM['segment'] = RFM['RF_SCORE'].replace(seg_map, regex=True)
```

```python
RFM.head()
```

🔴 Adım 1: Segmentlerin recency, frequnecy ve monetary ortalamalarını incele

```python
RFM.groupby("segment").agg({"Recency":"mean", "FREQUENCY": "mean", "MONETARY": "mean"})
```
